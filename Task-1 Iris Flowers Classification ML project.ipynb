```python
X = iris.iloc[:,0:4] 
y =iris.iloc[:,4]
```

### Split the data into training set and testing set using train_test_split
 

### Use train_test_split() with X and Y

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## fit a RandomForest model on the training set.

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print("Accuracy:",accuracy_score(y_pred,y_test))
```

## fit a Support Vector Meachine model on traing set. 

```python
model=SVC()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print("Accuracy:",accuracy_score(y_pred,y_test))
```

## fit a  Naive bayes model on traing set. 

```python
model=GaussianNB()
model.fit(X_train,y_train)     
```

```python
y_pred=model.predict(X_test)
print("Accuracy:",accuracy_score(y_pred,y_test))
```

## fit a  Decision tree model on traing set. 

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=1)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print("Accuracy:",accuracy_score(y_pred,y_test))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
classification_report(y_test, y_pred)
```
